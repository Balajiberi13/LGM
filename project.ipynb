### IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### IMPORTING DATSET

```python
columns=['Sepal length','Sepal width','Petal length','Petal width','Class_labels']
df=pd.read_csv("dataset.csv",names=columns)
df.head()
```

### VISUALISATION OF DATASET

```python
df.describe()
```

```python
sns.pairplot(df,hue='Class_labels')
```

### SEPERATING VECTOR MACHINE ALGORITHM

```python
data=df.values
X=data[:,0:4]
Y=data[:,4]
```

### SPLITTING THE DATA FOR TRAINING AND TESTING

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

### SUPPORT VECTOR MACHINE ALGORITHM

```python
from sklearn.svm import SVC
model_svc=SVC()
model_svc.fit(X_train,Y_train)
```

```python
prediction1=model_svc.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction1)*100)
```

### LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(X_train,Y_train)
```

```python
prediction2=model_LR.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction2)*100)
```

### DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
model_DTC=DecisionTreeClassifier()
model_DTC.fit(X_train,Y_train)
```

```python
prediction3=model_DTC.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction3)*100)
```

### CHECKING THE INPUT

```python
X_new=np.array([[3,2,1,0.2],[4.9,2.2,3.8,1.1],[5.3,2.5,4.9,1.9]])
prediction_res1=model_svc.predict(X_new)
prediction_res2=model_LR.predict(X_new)
prediction_res3=model_DTC.predict(X_new)

print("PREDICTION USING SUPPORT VECTOR MACHINE ALGORITHM :{}\n\nPREDICTION USING  LOGISTIC REGRESSION: {} \n\nPREDICTION USING DECISION TREE CLASSIFIER: {} ".format(prediction_res1,prediction_res2,prediction_res3))
```
