### IMPORT LIBRARY`

NAME: BALAJI BERI


```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

### READ IMAGE

```python
image=cv2.imread("after.jpg")
```

### TO SHOW OR DISPLAY IMAGE

```python
cv2.imshow("Balaji",image)
cv2.waitKey(0)
```

### READ IMAGE IN RGB FORMAT

```python
img_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

### CREATE GREYSCALE IMAGE

```python
grey_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
grey_image
```

```python
cv2.imshow("grey",grey_image)
cv2.waitKey(0)
plt.imshow(grey_image)
```

### INVERTING IMAGE

```python
inverted_image=255-grey_image
plt.imshow(inverted_image)
cv2.imshow("inverted",inverted_image)
cv2.waitKey(0)
```

### GAUSSIAN FUNCTION

```python
blurred=cv2.GaussianBlur(inverted_image,(21,21),0)
plt.imshow(blurred)
```

### INVERTING BLURRED

```python
inverted_blurred=255-blurred
pencil_sketch=cv2.divide(grey_image,inverted_blurred,scale=256.0)
plt.imshow(pencil_sketch)
```

 ### FINAL SKETCH

```python
cv2.imshow("Diksha Sketch",pencil_sketch)
cv2.waitKey(0)
```

### ORIGINAL IMAGE AND PENCIL SKETCH

```python
cv2.imshow("Original image",image)
cv2.imshow("pencil sketch",pencil_sketch)
cv2.waitKey(0)
```
